```python
import chess
import random

#read from file
f = open("all_moves_proba.txt", "r")
all_moves = {a:int(b) for a,b in [line.split(",") for line in f]}
print(len(all_moves))
```

```python

def simulate_game():
    seen_moves = set()
    board = chess.Board()
    print(board)
    i=0
    while not board.is_game_over():     
        i+=1 
        for move in board.legal_moves:
            seen_moves.add(board.san(move))
        board.push(random.choice(list(board.legal_moves)))
    print(i)
    return seen_moves

# generate random games and fetch all moves seen

for i in range(10000):
        seen_moves =simulate_game()
        for move in seen_moves:
            if move in all_moves:
                all_moves[move]+=1
            else:
                all_moves[move]=1
        print(len(all_moves))

print(all_moves)
```

```python
with open("all_moves_proba.txt", "w") as f:
    for val in all_moves:
        f.write(val+","+str(all_moves[val])+"\n")
```
